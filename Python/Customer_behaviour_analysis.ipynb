```python
#null values 

df.isnull().sum()
```

```python
#fill null values

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# columns in  lowercase and replace space with underscore 

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
# create a new column
labels = ['young adult','adult','middle-aged','senior']
df['age_group']= pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create purchase_frequency_days as frequency_of_purchases column contains text based data 

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' :90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
# Step 1: Install the required MySQL connector package
# Run this command first in a Jupyter cell:
!pip install mysql-connector-python

# Alternative option (if above doesn't work):
# !pip install PyMySQL

from sqlalchemy import create_engine

# Step 2: Connect to MySQL
# Replace placeholders with your actual details
username = "root"          # default MySQL user
password = "root"          # your MySQL password
host = "localhost"         # if running locally
port = "3306"              # default MySQL port
database = "customer_behavior" 

# Option 1: Using mysql-connector-python (after installation)
try:
    engine = create_engine(f"mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}")
    print("Using mysql-connector-python driver")
except Exception as e:
    print(f"mysql-connector failed: {e}")
    # Option 2: Fallback to PyMySQL (install with: !pip install PyMySQL)
    engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")
    print("Using PyMySQL driver")

# Step 3: Load DataFrame into MySQL
table_name = "customer"
# Note: Make sure 'df' variable is defined before running this
# df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Engine created successfully. Ready to load data into table '{table_name}' in database '{database}'.")
```
